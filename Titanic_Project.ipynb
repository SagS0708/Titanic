## **Correlation, ANOVA, CHIsquare**

```python
a.info()
```

```python
a["Survived"]=a["Survived"].astype(str)
```

```python
a["Survived"].replace("0","No",inplace=True)
a["Survived"].replace("1","Yes",inplace=True)
```

```python
a.head(3)
```

```python
def CHI_ANOVA(df,CAT_COL):
  def CATCON(df):
    cat=[]
    con=[]
    for i in df.columns:
      if df[i].dtypes=="object":
        cat.append(i)
      else:
        con.append(i)
    return cat,con
  cat,con=CATCON(df)
  cat.remove(CAT_COL)

  import pandas as pd
  imp_cat=[]
  imp_con=[]
  imp_col=[]
  #CHISquare
  from scipy.stats import chi2_contingency
  for value in cat:
    if chi2_contingency(pd.crosstab(df[CAT_COL],df[value]))[1] <=0.035:
      imp_cat.append(value)
    else:
      pass
  #ANOVA
  from statsmodels.formula.api import ols
  from statsmodels.stats.anova import anova_lm
  for value in con:
    rel = value + "~" + CAT_COL
    model = ols(rel,df).fit()
    results=anova_lm(model)
    if results.iloc[0,4] <=0.035:
      imp_con.append(value)
    else:
      pass
  #Important Columns
  imp_col = imp_cat + imp_con
  return imp_col
```

```python
imp_col=CHI_ANOVA(a,"Survived")
```

## **Preprocessing and Model Creation**

### **Model 1 (Best Model)**

```python
M=X[imp_col]

import pandas as pd
cat = []
con = []
for i in M.columns:
  if(M[i].dtypes == "object"):
    cat.append(i)
  else:
    con.append(i)
X1 = pd.get_dummies(M[cat])
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X2 = pd.DataFrame(ss.fit_transform(M[con]),columns=con)
Xnew = X2.join(X1)


from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(xtrain,ytrain)
pred=model.predict(xtest)

from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)

```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,pred)
```

```python
from sklearn.metrics import recall_score
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ytest1 = le.fit_transform(ytest)
pred1 = le.transform(pred)
```

```python
recall_score(ytest1,pred1)
```
